```python
from svc import SupportVectorClustering as SVC
import sklearn.datasets
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
def calc_and_print_result(p, q, x):
    svc = SVC(p, q)
    svc.fit(x)
    clusters = svc.get_clusters()
    print(f'p = {p}, q = {q}')
    print(f'clusters: {len(clusters)}')

    mar = ['x' if i in svc.svs_indx else 'P' if i in svc.bsv_indx else '.' for i in range(len(x))]
    cmap = plt.get_cmap("jet")
    color=  cmap(np.linspace(0, 1, len(clusters)))
    plt.figure() 
    for i in range(len(clusters)):  
        plt.scatter(x[:, 0][clusters[i]], x[:, 1][clusters[i]], color=color[i], marker=mar[i])


x,y = sklearn.datasets.make_blobs(n_samples=50, random_state=333, centers=3)
plt.figure()
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='brg')
```

```python
p, q = SVC.get_optimal_p(x), SVC.get_optimal_q(x)
svc = SVC(p, q)
svc.fit(x)
clusters = svc.get_clusters()
print(f'p = {p}, q = {q}')
print(f'clusters: {len(clusters)}')

col = [0] * len(x)
for key,value in clusters.items():
    for i in value:
        col[i] = key  

plt.scatter(x[:, 0], x[:, 1], c=col)
```

```python
scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)
p, q = SVC.get_optimal_p(x), SVC.get_optimal_q(x)
svc = SVC(p, q)
svc.fit(x)
clusters = svc.get_clusters()
print(f'p = {p}, q = {q}')
print(f'clusters: {len(clusters)}')

col = [0] * len(x)
for key,value in clusters.items():
    for i in value:
        col[i] = key  

plt.scatter(x[:, 0], x[:, 1], c=col)
x = scaler.inverse_transform(x)
```

```python
calc_and_print_result(0.02, 0.1, x=x)
```
